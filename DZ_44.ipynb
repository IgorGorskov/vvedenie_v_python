```python
import pandas as pd
import random
lst = ['robot'] * 10
print(lst)
lst += ['human'] * 10
print(lst)
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data
```

```python
# как я понял по условию,
# программа работает только для датафрейма в один столбец
# который создается в выше

my_lst = list()
lst_1 = list() #подсчет людей
lst_2 = list() #подсчет роботов
for i in range(data.shape[0]): #кол-во строк 
  for j in range(data.shape[1]): #кол-во столбцов
    if data.iloc[i][j] == 'human':
      lst_1.append('1')
      lst_2.append('0')
    elif data.iloc[i][j] == 'robot':
      lst_1.append('0')
      lst_2.append('1')
one_hot_data = pd.DataFrame({'IamHuman':lst_1,'IamRobot':lst_2})
data_combined = pd.concat([data, one_hot_data], axis=1)
data_combined 
```
